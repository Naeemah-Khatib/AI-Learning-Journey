```python
import pandas as pd
df = pd.read_excel("sales_data.xlsx")
```

```python
df
```

```python
def totalSales(x):
  return x["Sales"].sum()

totalSales(df)
```

```python
def totalCal(x):
  return x.sum()

totalCal(df["Sales"])
```

```python
totalSales(df.groupby("Region"))
```

```python
def top_product(df):

    result = (
        df.groupby("Product")["Sales"]
          .sum()
          .sort_values(ascending=False)
    )

    return result.head(1)

top_product(df)
```

```python
def executive_summary(df):

    total_sales_value = df["Sales"].sum()

    total_cost_value = df["Cost"].sum()

    print("Executive Summary")
    print("----------------")
    print("Total Sales:", total_sales_value)
    print("Total Cost:", total_cost_value)

executive_summary(df)
```

```python
def average_sales(df):
  x = df["Sales"].mean()
  return x

average_sales(df)
```

```python
def highest_sale(df):
  x = df["Sales"].sort_values(ascending=False).head(1)
  return x

print(highest_sale(df))

df[df['Sales']==df['Sales'].max()]
```

```python
def top_region(df):
  x = df.groupby('Region')["Sales"].sum()
  return x.head(1)

top_region(df)
```
