```python
import pandas as pd
from datetime import date
```

```python
def calc_total(col):
  return sum(col)

def calc_percentage(num,den):
  return round(num/den * 100,1)

def sales_by_item(df,name):
  return df.groupby(name)['Sales'].sum()
```

```python
def main(df_excel):

  #print(df)
  df_excel["Profit"] = df_excel["Sales"] - df_excel["Cost"]
  sales = calc_total(df_excel['Sales'])
  cost = calc_total(df_excel['Cost'])
  profit = calc_total(df_excel['Profit'])
  margin_perc = calc_percentage(profit,cost)

  summary = f"""
  ====================================
  SALES EXECUTIVE SUMMARY - {date.today()}
  ====================================

  Total Sales: {sales}
  Total Profit: {profit}
  Profit Margin: {margin_perc}%

  Top Region: {sales_by_item(df_excel,"Region").idxmax()}
  Top Product: {sales_by_item(df_excel,"Product").idxmax()}
  Top Customer: {sales_by_item(df_excel,"Customer").idxmax()}

  Sales by Region\n----------------
  {sales_by_item(df_excel,"Region")}
  """
  return summary
```

```python
df_excel = pd.read_excel("Project01_Sales_Data.xlsx")

output = main(df_excel)

#export as text
with open("ExecutiveSummary.txt","w") as file:
  file.write(output)
```

```python
#export to excel - add new sheet
new_df = sales_by_item(df_excel,"Region")
print(new_df)

with pd.ExcelWriter("New Excel.xlsx") as writer:
  new_df.to_excel(writer,sheet_name="Exec Summary", index=True)
```

```python
# better formatting
sales_by_region = df_excel.groupby("Region")["Sales"].sum()

for region, sales in sales_by_region.items():
    print(f"{region:<5} : ₹{sales:,.0f}")
```
