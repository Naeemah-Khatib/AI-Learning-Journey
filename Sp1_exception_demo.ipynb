```python

def calculate_average(numbers):
  try:
    sum = 0
    for item in numbers:
      sum += item
    return sum/len(numbers)

  except ZeroDivisionError:
    return None

  finally:
    print("Completed")
```

```python
res = calculate_average([10,20,30])
if res==None:
  print("Empty list")
else:
  print(res)
```

```python
def calculate_profit_margin(total_profit, total_sales):
    if not isinstance(total_profit, (int, float)):
        raise TypeError("total_profit must be a number.")

    if not isinstance(total_sales, (int, float)):
        raise TypeError("total_sales must be a number.")

    if total_sales == 0:
        raise ValueError("Total Sales cannot be zero.")

    if total_sales < 0:
        raise ValueError("Total Sales cannot be negative.")

    return (total_profit / total_sales) * 100
```
